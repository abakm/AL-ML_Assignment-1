# Data Preprocessing

```python
# Drop Duplicates
df.drop_duplicates(inplace=True)
df
```

```python
df.isnull().sum()  # checking for null values
```

```python
# Apply Label Encoding to sex, smoker and region

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

# Apply Label Encoding to the 'sex' column
df["sex"] = label_encoder.fit_transform(df["sex"])
df["smoker"] = label_encoder.fit_transform(df["smoker"])
df["region"] = label_encoder.fit_transform(df["region"])

df
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Apply MinMax Scaling to age and bmi
df_min_max_scaled = pd.DataFrame(scaler.fit_transform(df[["age", "bmi"]]))

# Assign column names to the scaled dataframe
df_min_max_scaled.columns = ["age_scaled", "bmi_scaled"]

# Method 1: Add the scaled columns to the original dataframe
df["age"] = df_min_max_scaled["age_scaled"]
df["bmi"] = df_min_max_scaled["bmi_scaled"]

df

```

## Model Development

```python
x = df[["age", "sex", "bmi", "children", "smoker", "region"]]
y = df["charges"]

```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```
