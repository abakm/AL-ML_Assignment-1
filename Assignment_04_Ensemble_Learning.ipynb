```python
# import statements
import numpy as np
import pandas as pd
import statistics as st
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split, StratifiedGroupKFold
from sklearn.ensemble import VotingClassifier, BaggingClassifier, AdaBoostClassifier, GradientBoostingClassifier
```

# Stacking

```python
from sklearn.model_selection import StratifiedKFold
```

```python
# Train first-level models and create meta-features
def stacking(model, train, y, test, n_fold):
  folds = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=1)
  train_meta_features, test_meta_features = np.zeros((train.shape[0], 1)), np.zeros((test.shape[0], 1))
  for train_idx, test_idx in folds.split(train, y):
    x_train, x_test = train.iloc[train_idx], train.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
    model.fit(x_train, y_train)
    predict = model.predict(x_test)
    test_predict = model.predict(test)
    train_meta_features[test_idx, 0] = predict
    test_meta_features[:, 0] += test_predict
  test_meta_features /= n_fold
  return train_meta_features, test_meta_features
```

### create two base models

```python
# create decision tree model
tree = DecisionTreeClassifier(random_state=1)
train_predict1, test_predict1 = stacking(model=tree, train=x_train, y=y_train, test=x_test, n_fold=10)
train_predict1 = pd.DataFrame(train_predict1)
test_predic1 = pd.DataFrame(test_predict1)
```

```python
# create KNN model
tree = KNeighborsClassifier()
train_predict2, test_predict2 = stacking(model=tree, train=x_train, y=y_train, test=x_test, n_fold=10)
train_predict2 = pd.DataFrame(train_predict2)
test_predict2 = pd.DataFrame(test_predict2)
```

### create meta model

```python
# create meta model ,fit the model and predict
meta_df = pd.concat([train_predict1, train_predict2], axis=1)
meta_df_test = pd.concat([test_predic1, test_predict2], axis=1)
lr = LogisticRegression()
lr.fit(meta_df, y_train)
lr.predict(meta_df_test)
```

```python
# score
lr.score(meta_df, y_train)
```

# Hyperparameter tuning

### using GridSearchCV

```python
parameters = {
    'n_estimators': [50, 100, 200],
    'max_depth': [10, 50, 60],
    "criterion": ["gini", "entropy"],
}

grid_search = GridSearchCV(RandomForestClassifier(), parameters, cv=5, scoring="accuracy", verbose=1, n_jobs=1)
grid_search.fit(x, y)

best_params = grid_search.best_params_
best_score = grid_search.best_score_

print("Best Parameters:", best_params)
print("Best Score:", best_score)
```

### using RandomizedSearchCV

```python
random_search = RandomizedSearchCV(RandomForestClassifier(), parameters, n_iter=18, cv=5, scoring="accuracy", verbose=1, n_jobs=1)
random_search.fit(x, y)

best_params = random_search.best_params_
best_score = random_search.best_score_

print("Best Parameters:", best_params)
print("Best Score:", best_score)
```
