```python
# import statements
import numpy as np
import pandas as pd
import statistics as st
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, StratifiedGroupKFold
from sklearn.ensemble import VotingClassifier, BaggingClassifier, AdaBoostClassifier, GradientBoostingClassifier
```

```python
# Read the dataset to the python environment.
df = pd.read_csv("./train_loan.csv")
df
```

### pre-processing steps.

```python
# check for null values

df.isnull().sum()
```

```python
# Remove  rows with null values

df.dropna(inplace=True)
df
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)
df
```

```python
# Remove unneccessary columns
df.drop(["Loan_ID"], axis=1, inplace=True)
df
```

```python

label_encoder = LabelEncoder()

# Apply Label Encoding to the 'Gender, Married, Education, Self_Employed and Property_Area columns
df["Gender"] = label_encoder.fit_transform(df["Gender"])
df["Married"] = label_encoder.fit_transform(df["Married"])
df["Education"] = label_encoder.fit_transform(df["Education"])
df["Self_Employed"] = label_encoder.fit_transform(df["Self_Employed"])
df["Property_Area"] = label_encoder.fit_transform(df["Property_Area"])

df
```

```python
# check for the values in the Dependents as its does have 3+ as values as its not a number so that we will preprocess them
df['Dependents'].unique()
```

```python
# Replace 3+ with 3

df['Dependents'] = pd.to_numeric(df['Dependents'].replace('3+', '3'), errors='coerce')
df['Dependents'] = df['Dependents'].astype(int)
df
```

#Perform different Ensemble Learning techniques in the given dataset.

```python
# split dataset
x =df.drop("Loan_Status", axis=1)
y = df["Loan_Status"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# create models



knc = KNeighborsClassifier()
dtc = DecisionTreeClassifier(random_state=1)
lrc = LogisticRegression()
```

```python
# Train Models
knc.fit(x_train, y_train)
dtc.fit(x_train, y_train)
lrc.fit(x_train, y_train)
```

```python
# Predict the models
print("KNN",knc.predict(x_test))
print("DT",dtc.predict(x_test))
print("LR",lrc.predict(x_test))
```

### Esemble

```python
final_predictions = np.array([])
for i in range(0, len(x_test)):
  final_predictions =  np.append(final_predictions, st.mode([knc.predict(x_test)[i], dtc.predict(x_test)[i], lrc.predict(x_test)[i]]))

final_predictions
```

### Max voting

```python
# Create models , combine model predictions
knc = KNeighborsClassifier()
dtc = DecisionTreeClassifier(random_state=1)
lrc = LogisticRegression()
vc = VotingClassifier(estimators=[('KNC', knc), ('DTC', dtc), ('LRC', lrc)], voting='hard')
vc.fit(x_train, y_train)
vc.score(x_test, y_test)
```

```python
### predict the model

vc.predict(x_test)
```

### Average Voting

```python
# create model
knc_avg = KNeighborsClassifier()
dtc_avg = DecisionTreeClassifier(random_state=1)
lrc_avg = LogisticRegression()
```

```python
# fit the models
knc_avg.fit(x_train, y_train)
dtc_avg.fit(x_train, y_train)
lrc_avg.fit(x_train, y_train)
```

```python
# predict the models
knc_avg_predict  = knc_avg.predict_proba(x_test)
dtc_avg_predict = dtc_avg.predict_proba(x_test)
lrc_avg_predict  = lrc_avg.predict_proba(x_test)
```

```python
# Final combined prediction
final_avg_prediction = (knc_avg_predict + dtc_avg_predict + lrc_avg_predict)/3
final_avg_prediction
```
